## タイトル・ラベル・凡例

### グラフタイトル
plt.title(タイトル名)

```python
x = ["朝","昼","夕","夜"]
y = [36.5,36.2,37.5,37.2]

plt.plot(x, y)
plt.title("2021-01-05の体温")
```

### 軸ラベル
plt.xlabel(横軸のラベル名)<br>
plt.ylabel(縦軸のラベル名)

```python
x = ["朝","昼","夕","夜"]
y = [36.5,36.2,37.5,37.2]

plt.plot(x, y)
plt.title("2021-01-05の体温")
plt.ylabel("体温(℃)")
```

### 凡例

plt.plot(x, y, label="ラベル名")<br>
plt.legend(loc="表示位置")

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, label="子")
plt.plot(x, y_f, label="父")
plt.plot(x, y_m, label="母")

plt.title("2021-01-05の体温")
plt.ylabel("体温(℃)")

plt.legend(loc="upper left")
```

### フォントサイズ
plt.title(title, fontsize=int)<br>
※タイトルの他に、ラベルや凡例のサイズも変更可能

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, label="子")
plt.plot(x, y_f, label="父")
plt.plot(x, y_m, label="母")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")
```

### グラフ色
plt.plot(x, y, color=color)<br>

引数のcolorはcと短縮も可能で、一部の色も短縮表記可能です。<br>
plt.plot(x, y, c="r")<br>

色の名前以外にも、カラーコードやRGB表記が可能。<br>

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, label="子", color="red")
plt.plot(x, y_f, label="父", color="orange")
plt.plot(x, y_m, label="母", color="blue")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")
```

### 透明度
plt.hist(data, alpha=0-1)<br>
ヒストグラムでよく用いられるが、他のグラフでも使える<br>

0~1の間で透明度を指定

```python
np.random.seed(250)
x1 = np.random.rand(50)
x2 = np.random.rand(30)

plt.hist(x1, alpha=0.5)
plt.hist(x2, alpha=0.5)
```

### 線のスタイル
plt.plot(x, y, linestyle="線の名前")<br>

linestyleはlsと省略可能

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, linestyle="solid", label="solid")
plt.plot(x, y_f, ls="--", label="dashed")
plt.plot(x, y_m, ls=":", label="dotted")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")
plt.legend()
```

### 線の太さ
plt.plot(x, y, linewidth=線の太さ)<br>

linewidthはlwと省略可能

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, linewidth=10, label="10 width")
plt.plot(x, y_f, lw=5, label="5 width")
plt.plot(x, y_m, lw=1, label="1 width")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")
plt.legend()
```

### マーカー
plt.scatter(x, y, marker=種類)<br>
折れ線グラフ・散布図でのマーカーの種類を変更する
<a href="https://matplotlib.org/3.2.0/gallery/lines_bars_and_markers/marker_reference.html#sphx-glr-gallery-lines-bars-and-markers-marker-reference-py">マーカーの種類</a>

```python
np.random.seed(250)
x1 = np.random.randn(50)
x2 = np.random.rand(30)

y1 = np.random.randn(50)
y2 = np.random.rand(30)

plt.scatter(x1, y1, marker="^")
plt.scatter(x2, y2, marker="x")
```

### グリッド
plt.grid()<br>

x軸もしくはy軸のみ線を引く場合には、引数にaxis="x"とする

```python
np.random.seed(250)
x1 = np.random.randn(50)
x2 = np.random.rand(30)

y1 = np.random.randn(50)
y2 = np.random.rand(30)

plt.scatter(x1, y1, marker="^")
plt.scatter(x2, y2, marker="x")
plt.grid()
```

### 補助線

plt.hlines(y座標, x座標の最小値（始点）, x座標の最大値（終点）)
plt.vlines(x座標, y座標の最小値（始点）, y座標の最大値（終点）)

```python
np.random.seed(250)
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, marker="^")

plt.hlines(y.mean(), x.max(), x.min(), ls="--", color="black")
plt.vlines(x.mean(), y.max(), y.min(), ls="--", color="black")
```

### 軸の表示領域
plt.xlim(横軸下限値,横軸上限値)<br>
plt.ylim(縦軸下限値,縦軸上限値)

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, label="子")
plt.plot(x, y_f, label="父")
plt.plot(x, y_m, label="母")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")

plt.xlim("朝","夜")
plt.ylim(35,38)
```

### グラフの保存
plt.savefig("ファイル名")<br>
引数でdpi=300など任意の解像度に可能。指定しない場合は100<br>
bbox_inches=tightで余白除去

```python
x = ["朝","昼","夕","夜"]
y_c = [36.5,36.2,37.5,37.2]
y_f = [36.2,36.5,36.8,36.3]
y_m = [35.8,35.9,36.2,36.5]

plt.plot(x, y_c, label="子")
plt.plot(x, y_f, label="父")
plt.plot(x, y_m, label="母")

plt.title("2021-01-05の体温", fontsize=20)
plt.ylabel("体温(℃)")

plt.xlim("朝","夜")
plt.ylim(35,38)
plt.savefig("../sample/test.png", dpi=300, bbox_inches="tight")
```
